```python
# abstraction of a class
class Animal:
    def speak(self):
        raise NotImplementedError("Subclasses must implement this method")
    
#inheritance
class Dog(Animal):
    def speak(self):
        return "Woof!"
    
class Cat(Animal):
    def speak(self):
        return "Meow!"
    
#polymorphism
def animal_sound(animal):
    print(animal.speak())

dog = Dog()
cat = Cat()

animal_sound(dog)  # Output: Woof!
animal_sound(cat)  # Output: Meow!  
```
